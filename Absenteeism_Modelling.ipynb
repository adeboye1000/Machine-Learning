```python
targets = targets.drop('Absenteeism Time in Hours', axis = 1)
targets
```

```python
unscaled_inputs = targets.iloc[:, :-1]
```

```python
unscaled_inputs.head()
```

```python
data_preprocessed['Excessive Absenteesm'] = targets
targets
```

```python
outputs = targets['Excessive Absenteesm']
outputs
```

```python
outputs.shape
```

# Standardize the inputs

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs.shape
```

# Split the Input data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(scaled_inputs, outputs, test_size= 0.2, random_state = 20)
```

```python
print(X_train.shape, X_test.shape)
```

# Training the model

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
```

```python
y_train.shape
```

```python
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

# Manually computing the accuracy

```python
model_outputs = reg.predict(X_train)
```

```python
model_outputs == y_train
```

```python
np.sum(model_outputs == y_train)
```

```python
model_outputs.shape[0]
```

```python
Accuracy = np.sum(model_outputs == y_train)/model_outputs.shape[0]
Accuracy
```

# Finding the intercepts and coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns = ['Feature name'], data = feature_name)
summary_table
```

```python
summary_table['Coefficients'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
```

```python
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
```

```python
summary_table = summary_table.sort_index()
```

```python
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficients)
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending = False)
```
